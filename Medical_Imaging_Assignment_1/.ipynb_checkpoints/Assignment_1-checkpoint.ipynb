```python
from matplotlib.colors import LogNorm
import matplotlib.pyplot as pl
import numpy as np
import phantominator as ph
from scipy.fft import fftshift
import scipy
```

# a) Load Shepp-Logan phantom

**Start by defining the numerical 2D Shepp-Logan phantom on a 128 x 128
matrix. Show the image in 2D**.

```python
#a
phantom = ph.shepp_logan(128)
pl.imshow(phantom)
pl.show()
```

# b) FFT

**Compute the 2D FFT of the numerical phantom. Show the real and
imaginary parts of the 2D FFT as images**.

```python
FT = np.fft.fft2(phantom)
FTS = fftshift(FT)

fig, ax = pl.subplots(1,3, figsize=(12, 4))
fig.suptitle("FFT of Shepp-Logan phantom")
ax[0].imshow(np.real(FTS), vmin = 0, vmax = 255)
ax[0].set_title("Real")
ax[1].imshow(np.imag(FTS), vmin = 0, vmax = 255)
ax[1].set_title("Imaginary")
ax[2].imshow(np.abs(FTS), vmin = 0, vmax = 255)
ax[2].set_title("Absolute value")

pl.show()
```

# c) Reducing the resolution

**Modify the Fourier data to reduce the resolution of the information by a
factor of 2, aiming for an image on a 64 x 64 matrix. Hint: you will need
to work out how much of the Fourier information needs to be eliminated.
Show the real and imaginary parts of the modified version of the 2D FFT
as ima**

## Solution
1. Generate a filter with values between 0 and 1
    - For box filter, a square of 64x64 has values of 1, and every other pixel has value of 0
    - For 2D gaussian filter, the standard deviation is 32 and it is centered in the middle of the image
2. Multiply the filter with the FT (i.e. cut off the highest frequencies in Fourier space)
3. Plot to compare real and imaginary partsges.

```python
# Removing the highest 1/2 of frequencies in fourier domain (box filter)
size_filter = 64
ones = np.ones((size_filter, size_filter))
filter = np.pad(ones, (np.shape(FTS)[0] - size_filter)//2, constant_values=0)
low_freq = filter*FTS

fig, ax = pl.subplots(1,2, figsize=(9, 4))
fig.suptitle("FFT with box filter")
ax[0].imshow(np.real(low_freq), vmin = 0, vmax = 255)
ax[0].set_title("Real")
ax[1].imshow(np.imag(low_freq), vmin = 0, vmax = 255)
ax[1].set_title("Imaginary")

pl.show()

# Another method: gaussian filter
# definition of 2d gaussian: https://stackoverflow.com/a/56923189/25652975
num_samples = 128
def gaus2d(x=0, y=0, mx=0, my=0, sx=32, sy=32):
    return np.exp(-((x - mx)**2. / (2. * sx**2.) + (y - my)**2. / (2. * sy**2.)))

x = np.arange(0, 128)
y = np.arange(0, 128)
x, y = np.meshgrid(x, y)
gaussian_filter = gaus2d(x, y, mx=64, my=64) # max of 1
low_freq_gauss = FTS*gaussian_filter

fig, ax = pl.subplots(1,2, figsize=(9, 4))
fig.suptitle("FFT with Gaussian filter")
ax[0].imshow(np.real(low_freq_gauss), vmin = 0, vmax = 255)
ax[0].set_title("Real")
ax[1].imshow(np.imag(low_freq_gauss), vmin = 0, vmax = 255)
ax[1].set_title("Imaginary")

pl.show()
```

# d) Inverse FFT of modified Fourier data

image**.

##  Do you need to show the real or imaginary part, or something else

The absolute value should be used (see comparison plots below), because it encompasses the most information and represents the signal that would be detected in a real-world system, which typically would detect the amplitude of the intensity. The imaginary portion of the IFFT seems to be mostly noise for both methods.  

## What "features" do you observe in the resampled image?

In both resamplings, it is clear that some of the detailed features are lost. This is most apparent along the edge of the cranium and other large structures, where the edges are blurred compared to the original 128x128 image. In the box-filtered image, we also see some ripples in the solid-colored portions, indicating some aliasing did occur, but these are absent from the gaussian-filtered image.

e?

```python
# We can now IFFT and downsample the phantom without being afraid of aliasing
IFT = np.fft.ifft2(low_freq)
low_res = IFT[::2, ::2]

# plotting IFFT for box filter
fig, ax = pl.subplots(1,3, figsize=(12, 4))
fig.suptitle("Low-res Shepp-Logan phantom (box filter)")
ax[0].imshow(np.real(low_res))
ax[0].set_title("Real")
ax[1].imshow(np.imag(low_res))
ax[1].set_title("Imaginary")
ax[2].imshow(np.abs(low_res))
ax[2].set_title("Absolute value")

pl.show()
```

```python
# same for gaussian filtering
IFT = np.fft.ifft2(low_freq_gauss)
low_res_g = IFT[::2, ::2]

fig, ax = pl.subplots(1,3, figsize=(12, 4))
fig.suptitle("Low-res Shepp-Logan phantom (Gaussian filter)")
ax[0].imshow(np.real(low_res_g))
ax[0].set_title("Real")
ax[1].imshow(np.imag(low_res_g))
ax[1].set_title("Imaginary")
ax[2].imshow(np.abs(low_res_g))
ax[2].set_title("Absolute value")

pl.show()
```

# e) Adding filtering before FFT

Repeat the downsampling exercise, but this time by filtering the numerical 
phantom image first. Apply a Gaussian filter to the image generated i 
step (a), before re-doing the 2D FFT, downsampling, and inverse 2D F T
described in steps (b), (c), and (d). Show the new downsampled ima e.
Explore different values for the width (sigma) of the filter (do not con use
this with the size of the filter) to see what value results in approp iate
filt

By inspection, a width of about 1 looks best (see figures below).ing.**

```python
def gaussian_filter_downsampling(sigma):
    
    # Apply Gaussian filter
    blurred_phantom = scipy.ndimage.gaussian_filter(phantom, sigma)
    # FT
    FT = np.fft.fft2(blurred_phantom)
    FTS = fftshift(FT)
    
    # truncate high frequencies
    gaussian_filter = gaus2d(x, y, mx=64, my=64)
    low_freq_gauss = FTS*gaussian_filter
    
    # Inverse FT
    IFT = np.fft.ifft2(low_freq_gauss)
    return IFT[::2, ::2]

s = [0.5, 1, 3, 10]
fig, ax = pl.subplots(1,len(s), figsize=(12, 4))
fig.suptitle("Comparing widths of Gaussian kernel and their effect on downsampled image")

for i in range(len(s)):
    low_res_filtered = gaussian_filter_downsampling(s[i])
    ax[i].imshow(np.abs(low_res_filtered))
    ax[i].set_title(r"$\sigma$ = {}".format(s[i]))

pl.show()
```
